```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("spambase.csv")
```

```python
df.head()
```

```python
predict = 'spam'
X = np.array(df.drop([predict], axis=1))
Y = np.array(df[predict])
```

```python
df.describe().T
```

```python
from scipy.stats import zscore

df_zscore = df.apply(zscore)
df_zscore
print("original df:", df.shape)
```

```python
outliers = df_zscore[(df_zscore.abs() > 3).any(axis=1)]
print(outliers)
```

```python
df = df.drop(index=outliers.index)
print("df after deleting noises:", df.shape)
```

```python
from sklearn import preprocessing

normalized_df = preprocessing.normalize(df)
print(normalized_df)
x = normalized_df
```

```python
from sklearn.linear_model import LogisticRegression

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
mymodel = LogisticRegression(solver='liblinear', max_iter=1000)
mymodel.fit(x_train, y_train)
```

# Accuracy

```python
from sklearn.metrics import accuracy_score

y_pred = mymodel.predict(x_test)
accu = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accu}")
```

# Preceision 

```python
from sklearn.metrics import precision_score

pre = precision_score(y_test, y_pred)
print(f"Precision: {pre}")
```

# Recall

```python
from sklearn.metrics import recall_score
 
recall = recall_score(y_test, y_pred)
print(f"Recall: {recall}") 
```

# F1 Score

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)
print(f"F1 score: {f1}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

# Using L1

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
mymodel = LogisticRegression(penalty='l1',solver='liblinear', max_iter=1000)
mymodel.fit(x_train, y_train)
```

```python
y_pred = mymodel.predict(x_test)
accu = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accu}")
```

# Using L2 

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
mymodel = LogisticRegression(penalty='l2',solver='liblinear', max_iter=1000)
mymodel.fit(x_train, y_train)
```

```python
y_pred = mymodel.predict(x_test)
accu = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accu}")
```
